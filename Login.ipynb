```python
#Importting libraries
import hashlib
import re
import os
errorflag = False


#This function check if password is valid.
def valid_password(password):
  pattern_pwd = "^.*(?=.{6,15})(?=.*\d)(?=.*[a-z])(?=.*[A-Z])(?=.*[@#$%^&+=]).*$"
  result = re.findall(pattern_pwd, password)
  if(result):
    return(True)
  else:
    return(False)


#This function check if Email ID is valid.
def valid_email(email_id):
  pattern_email = r'\b[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.com'
  if(re.fullmatch(pattern_email, email_id)):
    i1 = int((re.search('@', email_id)).start())
    i2 = int((re.search('\.', email_id)).start())
    if((i2-i1 > 0) & (i1 > 0)):
      return(True)
    else:
      return(False)
  else:
    return(False)
  

#This function handles "Forgot Password" functionality.
def forgotpassword():
  loginFlag = False
  email = input("Enter email address: ")
  if(valid_email(email)):
    if (os.path.isfile("Login_Register.txt")):
      with open("Login_Register.txt", "r") as f:
        f.seek(0)
        if(f.readlines() != []):
          for line in f:
            user_details = line.split(' ')
            stored_email = user_details[0]
            stored_pwd = user_details[1]
            if email == stored_email:
              pwd = stored_pwd.encode()
              pwd_hash = hashlib.md5(pwd).hexdigest()
              print("Password is ", pwd_hash)
              loginFlag = True
          if(loginFlag == False):
            print("Email ID is not registered. Choice Register to create an account \n\n")
        else:
          print("Login file is empty\n")
      f.close()
    else:
      print("Login file does not exist\n")
  else:
    print("Invalid email ID. Enter a Valid Emil ID")


#This function handles "Sign in", thereby making an entry in the text file.
def register():
    email = input("Enter email address: ")
    pwd = input("Enter password: ")
    conf_pwd = input("Confirm password: ")
    if valid_email(email):
      if(valid_password(pwd)):
        if conf_pwd == pwd:
          enc = conf_pwd.encode()
          hash1 = hashlib.md5(enc).hexdigest()
          with open("Login_Register.txt", "a+") as f:
             f.write(email + ' ')
             f.write(hash1 + '\n')
          f.close()
          print("You have registered successfully!\n\n")
        else:
          print("Passwords does not match ! \n")
      else:
        print("Weak Password. Enter Strong password\n")
    else:
      print("Email ID is not valid\n")


#This function check for the entry in text file to login in.
def login():
  loginFlag = False
  email = input("Enter email: ")
  pwd = input("Enter password: ")
  auth = pwd.encode()
  auth_hash = hashlib.md5(auth).hexdigest()
  if (os.path.isfile("Login_Register.txt")):
    with open("Login_Register.txt", "r") as f:
      f.seek(0)
      if(f.readlines() != []):
        for line in f:
          user_details = line.split(' ')
          stored_email = user_details[0]
          stored_pwd = user_details[1]
          if email == stored_email and auth_hash == stored_pwd:
            print("Logged in Successfully!\n\n")
            loginFlag = True
            break
        if(loginFlag != False):
          print("Login failed! \n")
          errorflag = True
      else:
        print("Login file is empty\n")
    f.close()
  else:
    print("Login file does not exist\n")


# <<< Main code begins >>       
while 1:
    print("********** Login System **********")
    print("\n\n1.Register")
    print("2.Login")
    print("3.Exit")
    if(errorflag):
      print("4.Forgot Password")
      errorflag = False

    ch = int(input("Enter your choice: "))
    if ch == 1:
        register()
    elif ch == 2:
        login()
    elif ch == 3:
        break
    elif ch == 4:
        forgotpassword()
    else:
        print("Wrong Choice!")
```
